Vamos fazer um exercício completo de pandas para um miniprojeto de análise de dados.

Esse exercício vai obrigar a gente a usar boa parte dos conhecimento de pandas e até de outros módulos que já aprendemos ao longo do curso.

### O que temos?

Temos os dados de 2019 de uma empresa de prestação de serviços. 

- CadastroFuncionarios
- CadastroClientes
- BaseServiçosPrestados

Obs1: Para ler arquivos csv, temos o read_csv<br>
Obs2: Para ler arquivos xlsx (arquivos em excel normais, que não são padrão csv), temos o read_excel

### O que queremos saber/fazer?

1. Valor Total da Folha Salarial -> Qual foi o gasto total com salários de funcionários pela empresa? <br>
    Sugestão: calcule o salário total de cada funcionário, salário + benefícios + impostos, depois some todos os salários
    
    
2. Qual foi o faturamento da empresa?<br>
    Sugestão: calcule o faturamento total de cada serviço e depois some o faturamento de todos
    
    
3. Qual o % de funcionários que já fechou algum contrato?<br>
    Sugestão: na base de serviços temos o funcionário que fechou cada serviço. Mas nem todos os funcionários que a empresa tem já fecharam algum serviço.<br>
    . Na base de funcionários temos uma lista com todos os funcionários<br>
    . Queremos calcular Qtde_Funcionarios_Fecharam_Serviço / Qtde_Funcionários_Totais<br>
    . Para calcular a qtde de funcionários que fecharam algum serviço, use a base de serviços e conte quantos funcionários tem ali. Mas lembre-se, cada funcionário só pode ser contado uma única vez.<br><br>
    Dica: se você aplicar o método .unique() em uma variável que é apenas 1 coluna de um dataframe, ele vai excluir todos os valores duplicados daquela coluna.<br>
    Ex: unicos_colunaA = dataframe['colunaA'].unique() te dá como resposta uma lista com todos os itens da colunaA aparecendo uma única vez. Todos os valores repetidos da colunaA são excluidos da variável unicos_colunaA 
    
    
4. Calcule o total de contratos que cada área da empresa já fechou


5. Calcule o total de funcionários por área


6. Qual o ticket médio mensal (faturamento médio mensal) dos contratos?<br>
    Dica: .mean() calcula a média -> exemplo: media_colunaA = dataframe['colunaA'].mean()

Obs: Lembrando as opções mais usuais de encoding:<br>
encoding='latin1', encoding='ISO-8859-1', encoding='utf-8' ou então encoding='cp1252'

Observação Importante: Se o seu código der um erro na hora de importar os arquivos:<br>
- CadastroClientes.csv
- CadastroFuncionarios.csv

Use separador ";" (ponto e vírgula) para resolver

```python
import pandas as pd

base_servico_df = pd.read_excel('BaseServiçosPrestados.xlsx')
cad_clientes = pd.read_csv('CadastroClientes.csv', sep=';', decimal=',')
cad_fun = pd.read_csv('CadastroFuncionarios.csv', sep=';', decimal=',')

display(base_servico_df)
```

### Primeira tarefa

- Descobrir a folha salarial dos funcionarios da empresa

```python
display(cad_fun)
salario_base = cad_fun['Salario Base'].sum()
impostos = cad_fun['Impostos'].sum()
beneficios = cad_fun['Beneficios'].sum()

print('R${:,.2f} gastos em pagamento salarial'.format(salario_base).replace(',','_').replace('.', ',').replace('_', '.'))
print('R${:,.2f} gastos em impostos'.format(impostos).replace(',','_').replace('.', ',').replace('_', '.'))
print('R${:,.2f} gastos em beneficios do funcionario'.format(beneficios).replace(',','_').replace('.', ',').replace('_', '.'))

folha_salarial = salario_base + impostos + beneficios
print('A folha salarial da empresa foi de: R${:,.2f}'.format(folha_salarial).replace(',','_').replace('.', ',').replace('_', '.'))
```

### Segunda Tarefa

- Faturamento da Empresa

```python
cad_clientes = cad_clientes[['ID Cliente', 'Valor Contrato Mensal']]
display(cad_clientes)


cad_clientes_2 = cad_clientes.merge(base_servico_df, on='ID Cliente')
display(cad_clientes_2)
```

```python
valor_mensal = cad_clientes_2['Valor Contrato Mensal'] * cad_clientes_2['Tempo Total de Contrato (Meses)']
print('A empresa faturou R${:,.2f}'.format(valor_mensal.sum()).replace(',','_').replace('.', ',').replace('_', '.'))
```

### Terceira Tarefa

- % de funcionarios que fecharam contrato

```python
cad_fun_2 = cad_fun[['ID Funcionário', 'Nome Completo']]
# display(cad_fun_2)

## juntando data frames
cad_fun_2 = cad_fun_2.merge(base_servico_df[['ID Funcionário','ID Cliente']], on='ID Funcionário')
display(cad_fun_2)

cad_fun_2.to_csv('Cad_fun_2.csv', sep=';', index=None)

total_fecharam_contrato = len(cad_fun_2['ID Funcionário'].unique())
print(total_fecharam_contrato)

total_fun = len(cad_fun['ID Funcionário'].unique())
print(total_fun)

porcentagem = total_fecharam_contrato / total_fun

display('{:.2%} dos funcionários fecharam contrato.'.format(porcentagem))
```

### Quarta Tarefa

- total de contratos que cada área da empresa já fechou

```python
base_servico_df = pd.read_excel('BaseServiçosPrestados.xlsx')
cad_clientes = pd.read_csv('CadastroClientes.csv', sep=';', decimal=',')
cad_fun = pd.read_csv('CadastroFuncionarios.csv', sep=';', decimal=',')

base_servico_df = base_servico_df[['ID Funcionário', 'ID Cliente', 'Tempo Total de Contrato (Meses)']]
cad_clientes = cad_clientes[['ID Cliente', 'Cliente', 'Valor Contrato Mensal']]
cad_fun = cad_fun[['ID Funcionário', 'Salario Base', 'Impostos', 'Beneficios', 'Area']]

base_servico_df = base_servico_df.merge(cad_clientes, on='ID Cliente')
base_servico_df = base_servico_df.merge(cad_fun, on='ID Funcionário')

area = base_servico_df['Area'].unique()
print('Áreas da empresa: {}'.format(area))

print('='*35)

area_Financeiro = len(base_servico_df[base_servico_df['Area'] == 'Financeiro'])
print('A área Financeira fechou {} contratos'.format(area_Financeiro))

print('='*35)

area_Logística = len(base_servico_df[base_servico_df['Area'] == 'Logística'])
print('A área de Logística fechou {} contratos'.format(area_Logística))

print('='*35)

area_Comercial = len(base_servico_df[base_servico_df['Area'] == 'Comercial'])
print('A área Comercial fechou {} contratos'.format(area_Comercial))

print('='*35)

area_Administrativo = len(base_servico_df[base_servico_df['Area'] == 'Administrativo'])
print('A área Administrativa fechou {} contratos'.format(area_Administrativo))

print('='*35)

area_Operações = len(base_servico_df[base_servico_df['Area'] == 'Operações'])
print('A área de Operações fechou {} contratos'.format(area_Operações))

print('='*35)

print('Total de contratos: {}'.format(area_Operações + area_Administrativo + area_Comercial + area_Logística + area_Financeiro))
```
